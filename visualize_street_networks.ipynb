```python
import osmnx as ox
from IPython.display import Image
```

```python
# Define cities list to be plotted
cities = {
    "San Francisco": (37.793897, -122.402189),
    "Boston": (42.35898789630188, -71.05962150203321),  # Haymarket 
    "New York City": (40.758872403952275, -73.97849752620358),   # 30 Rock
    "Paris": (48.86469016268871, 2.334908690146568), # Palais-Royal
    "Amsterdam": (52.372178543506124, 4.891322037872843),    # Kononklijk Paleis
    "Tokyo": (35.68119479928526, 139.76732374234209),    # Tokyo Station
}
```

```python
def plot_point(place, point, network_type="drive", default_width=2):
    img_folder = "images"
    extension = "png"
    dpi = 200
    fp = f"./{img_folder}/{place}.{extension}"
    G = ox.graph.graph_from_point(
        point, dist=1000, network_type=network_type, truncate_by_edge=True
    )
    fig, ax = ox.plot.plot_figure_ground(
        G=G,
        dist=1000,
        default_width=default_width,
        bgcolor='w',
        color='k',  # plot_graph takes "edge_color" but plot_figure_to_graph takes "color"
        edge_linewidth=0.2,
        filepath=fp,
        dpi=dpi,
        save=True,
        show=False,
        close=True,
    )
    return fp
```

```python
city = 'New York City'
fp = plot_point(city, cities[city], "drive")
Image(fp)
```
